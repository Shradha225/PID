```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
import control as ctrl
```

```python
num = np.array([1])
den = np.array([1,3,5,1])
G= ctrl.tf(num,den)
T=G/(1+G)
print("h is ",T)
t,y =ctrl.step_response(T*2) #scaled 
noise_amplitude = 0.05  
noise = noise_amplitude * np.random.normal(0, 1, len(y))

# Noisy output signal
y_noisy = y + noise

# Plot the results
plt.figure(figsize=(8, 5))
plt.plot(t, y, label="Ideal Response", linewidth=2)
plt.plot(t, y_noisy, label="Noisy Response", linestyle='--', color='r', alpha=0.7)
plt.xlabel("Time (s)")
plt.ylabel("Output")
plt.title("Step Response with Sensor Noise")
plt.legend()
plt.grid(True)
plt.show()
```
